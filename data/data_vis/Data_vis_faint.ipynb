```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv('../../data/trian_qso_wo_petro_v2.csv')
```

```python
df = data[['u - g', 'u - r', 'u - i', 'u - z', 'u - j', 'u - k',
       'u - ch1', 'u - ch2', 'g - r', 'g - i', 'g - z', 'g - k', 'g - j',
       'g - ch1', 'g - ch2', 'r - i', 'r - z', 'r - j', 'r - ch1', 'r - k',
       'r - ch2', 'i - z', 'i - j', 'i - k', 'i - ch1', 'i - ch2', 'z - j',
       'z - k', 'z - ch1', 'z - ch2', 'j - ch1', 'j - ch2', 'j - k', 'k - ch1',
       'k - ch2', 'ch1 - ch2','class_bin']]
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df.iloc[:,11:14], 
               c= df['class_bin'],
               figsize=(15,11));

scatter_matrix(df.iloc[:,19:22], 
               c= df['class_bin'],
               figsize=(15,11));
```

```python
from scipy.spatial import ConvexHull

plt.clf()
plt.figure(figsize = (10, 6))
names = ['non_quasar','quasar']
colors = ['b','r']
label = df['class_bin'].astype(np.int)
plt.xlabel('g - j', fontsize = 18)
plt.ylabel('j - k', fontsize = 18)

for i in range(len(names)):
    bucket = df[df['class_bin'] == i]
    bucket = bucket.iloc[:,[12,32]].values
    hull = ConvexHull(bucket)
    plt.scatter(bucket[:, 0], bucket[:, 1], label=names[i]) 
    for j in hull.simplices:
        plt.plot(bucket[j,0], bucket[j,1], colors[i])

plt.legend()
plt.savefig('../../figures/faint/gjk.png')
plt.show()
```
