```python
import pandas as pd
```

```python
df = pd.read_csv('../data/raw/S82X_galaxies_c2.csv')
```

```python
df = df[df['CH1_SPIES'] != -99.0].drop(columns = ['CH1_SHELA','CH2_SHELA','W1'])
df = df[df['G_mag'] != -99.000000]
df = df[df['U_mag'] != -99.000000]
df = df[df['I_mag'] != -99.000000]
df = df[df['Z_mag'] != -99.000000]
df = df[df['J_mag'] != -99.000000]
df = df[df['H_mag'] != -99.000000]
df = df[df['K_mag'] != -99.000000]
```

```python
df.rename(columns = {'CH1_SPIES':'W1mag','CH2_SPIES':'W2mag','SPEC_CLASS':'class'},inplace = True)
```

```python
df['class'] = 'faint_Galaxy'
```

```python
df.describe()
```

```python
faint = df[df['G_mag'] >= 22]
faint = faint[faint['R_mag'] >= 22]
faint
```

```python
df.to_csv('../data/faint_galaxy_set2.csv')
```
