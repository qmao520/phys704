```python
from astroquery.sdss import SDSS
from astropy import coordinates as coords
from astropy.coordinates import SkyCoord
from astropy.io import ascii
from astropy import units as u
```

```python
galaxy_table = ascii.read("SDSS_comparison_galaxies.csv",format='csv')
ra = galaxy_table['ra']
dec = galaxy_table['dec']

num = ra.size
```

```python
for ii in range(num):
    pos = SkyCoord(ra=ra[ii]*u.degree,dec=dec[ii]*u.degree)
    try: 
        xid = SDSS.query_region(pos,radius=1.0*u.arcsec,photoobj_fields= ['ra','dec','petroR50_g','petroR50_r','petroR50_i']) 
    except ValueError: 
        xid = None
        pass
    print(xid)
```
